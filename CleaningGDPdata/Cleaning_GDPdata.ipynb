```python
import pandas as pd
```

```python
# careful the csvpath if you try to run it
csvpath = 'SNA_TABLE1_09032023223137263.csv'
data = pd.read_csv(csvpath)
df = pd.DataFrame(data)
```

```python
B1_GA_data = df.loc[df['TRANSACT'] == 'B1_GA']
C_B1_GA_data = B1_GA_data.loc[B1_GA_data['MEASURE'] == 'C']
C_B1_GA_data.head()
```

```python
cleaned_data = C_B1_GA_data[['Country','Year','Unit','Value']]
cleaned_data.head(10)
```

```python
cleaned_data = cleaned_data.reset_index(drop = True)
df['Year'] = df['Year'].astype(int)
# first version cleaned data 
cleaned_data.to_csv("Cleaned_GDP_with2021&2022.csv")
# some country dont have data of 2022 and 2021
# in case some bugs so I delete 2022 and 2021
cleaned_data = cleaned_data.loc[cleaned_data['Year'] < 2021, :]
cleaned_data.head(15)
```
