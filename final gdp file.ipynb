```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Gross domestic product (GDP) is a monetary measure of the market value of all final goods and services produced in a period (quarterly or yearly) of time. GDP gives a rough estimate of the wealth and prosperity of the nation. It is considered as an important factor for any nation as it roughly correlates with a nation's life expectancy,average income of a person,amount of energy that nation consumes,and the average level of education of its citizens. Below is a brief analysis of the world gdp.

```python
# Reading csv for GDP
df_GDP = pd.read_csv('NEWDATA (2).csv')
df_GDP
```

```python
# Checking the number of countries.
num_country = df_GDP["Country"].nunique()
num_country
```

# Top Ten poorest countries

```python
# sorting the values for nominal GDP
Poorcountries_df=df_GDP.sort_values(["nominal GDP" ], ascending=True)
Poorcountries_df.head()
```

```python
#comparision of Poorest countries with nominal GDP
countries = Poorcountries_df.iloc[0:10]
countries= countries [["Country","nominal GDP"]]
countries =countries .groupby('Country').sum()
countries
```

# Plotting Bar chats

```python
#plotting on bar chart
plot_pandas = countries.plot.bar(color='b')  
plt.xlabel("Country")
plt.ylabel("nominal GDP")
plt.title("GDP")
plt.xticks(rotation="45")
```

```python
#comparision of Ten poorest countries with GDP per captial
countries = Poorcountries_df.iloc[0:10]
countries= countries [["Country","GDP per captial"]]
countries =countries .groupby('Country').sum()
countries
```

```python
#plotting on bar chart
plot_pandas = countries.plot.bar(color='b')  
plt.xlabel("country")
plt.ylabel("GDP per captial")
plt.title("GDP")
plt.xticks(rotation="45")
```

```python
#comparision of Poorest countries Population
countries = Poorcountries_df.iloc[0:10]
countries= countries [["Country","Population"]]
countries =countries .groupby('Country').sum()
countries
```

```python
#plotting on bar chart
plot_pandas = countries.plot.bar(color='b')  
plt.xlabel("country")
plt.ylabel("Population")
plt.title("GDP")
plt.xticks(rotation="45")
```

# Summary statistics

```python
mean = df_GDP['Population'].groupby(df_GDP['Continents']).mean()
```

```python
#Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of POPULATIONfor each Continent
mean = df_GDP['Population'].groupby(df_GDP['Continents']).mean()
median = df_GDP['Population'].groupby(df_GDP['Continents']).median()
var = df_GDP['Population'].groupby(df_GDP['Continents']).var()
std = df_GDP['Population'].groupby(df_GDP['Continents']).std()
sem = df_GDP['Population'].groupby(df_GDP['Continents']).sem()
summary_stat = pd.DataFrame({"mean Population":mean, 
                            "median Population":median, 
                           "Population Variance":var, 
                           "Population Std. Dev.":std, 
                           "Population Std. Err.":sem})
# Display the Summary statistics
summary_stat.head()
```

# Scatter plots

```python
# Build scatter plot for lat vs. lng
plt.scatter(     
    df_GDP["Lat"],
   df_GDP ["Lng"],
    edgecolor="black",
    linewidths=1,
    marker="o", 
    alpha=0.8,
    label="Country"
)

# Incorporate the other graph properties
plt.title("GDP")
plt.xlabel("Lat")
plt.ylabel("lng")
plt.grid(True)


# Show plot
plt.show()
```

```python
# Build scatter plot for Share of world GDP VS. GDP growth
plt.scatter(     
    df_GDP["Share of World GDP"],
   df_GDP ["GDP growth"],
    edgecolor="black",
    linewidths=1,
    marker="o", 
    alpha=0.8,
    label="GDP"
)

# Incorporate the other graph properties
plt.title(" GDP")
plt.xlabel("GDP growth")
plt.ylabel("Share of World GDP")
plt.grid(True)


# Show plot
plt.show()
```
